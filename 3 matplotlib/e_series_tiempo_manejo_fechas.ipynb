```python
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import pandas as pd
import numpy as np
```

```python
dates = pd.date_range(start='2023-01-01', periods=100)

values = np.random.rand(100).cumsum()

data = pd.DataFrame({'Date': dates, 'Values': values})
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(data['Date'], data['Values'], color='green')

plt.xticks(rotation=45)

plt.title('Serie de tiempo con formato de fechas')
plt.xlabel('Date')
plt.ylabel('Value')

plt.show()
```

```python
# graficar las ventas mensuales, manejando series de tiempo

dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales = np.random.randint(1000, 5000, size=12)

sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

plt.plot(sales_data['Date'], sales_data['Sales'], 
        marker='o',
        linestyle='-',
        label='Ventas Mensuales')

plt.gca().xaxis.set_major_formatter(DateFormatter('%b-%Y'))

plt.xticks(rotation=45)

plt.title('Analisis de Ventas Mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()
plt.tight_layout()

plt.show()
```
