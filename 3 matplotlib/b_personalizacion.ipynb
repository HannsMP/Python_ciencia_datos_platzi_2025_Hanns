```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores_studnet_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_studnet_2 = [50, 58, 63, 69, 74, 78, 83]
```

```python
plt.figure(figsize=(8,6))

plt.scatter(hours, exam_scores_studnet_1, 
    marker='o', 
    color='green',
    label='Estudiante 1'
)

plt.scatter(hours, exam_scores_studnet_2, 
    marker='s', 
    color='red',
    label='Estudiante 2'
)
plt.title('Relacion entre horas estudiadas y el puntaje de dos estudiantes')
plt.xlabel('Horas')
plt.xticks(rotation=45)
plt.ylabel('Puntaje')

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.plot(hours, exam_scores_studnet_1, 
    marker='o', 
    color='green', 
    linestyle='-',
    linewidth=2,
    label='Estudiante 1'
)

plt.plot(hours, exam_scores_studnet_2, 
    marker='s', 
    color='red',
    linestyle='--',
    linewidth=2,
    label='Estudiante 2'
)
plt.title('Relacion entre horas estudiadas y el puntaje de dos estudiantes')
plt.xlabel('Horas')
plt.xticks(rotation=45)
plt.ylabel('Puntaje')

plt.show()
```
