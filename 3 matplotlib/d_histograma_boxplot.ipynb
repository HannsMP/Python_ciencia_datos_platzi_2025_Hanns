```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.normal(170, 10, 200)

print(data)
```

```python
plt.hist(data, 
        color='salmon', 
        edgecolor='black', 
        bins=20,
        alpha=.6)

plt.title('Distribucion de altura')
plt.xlabel('Altura(cm)')
plt.ylabel('densidad')

plt.show()
```

```python
np.random.seed(0)

ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]

print(ages)
```

```python
plt.boxplot(ages,
            patch_artist=True,
            notch=True,
            vert=True,
            label=['Grupo1', 'Grupo2', 'Grupo3'])

plt.title('Distribucion de Edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad(Aos)')

plt.show()
```

```python
# Datos aleatorios
np.random.seed(0)
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# Histogram 2D
plt.hist2d(x, y, bins=30, cmap='Blues')
plt.colorbar()
plt.show()
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar()
plt.show()
```
